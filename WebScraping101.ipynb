```
import requests
from bs4 import BeautifulSoup
```

```
r = requests.get(r"https://www.seriouseats.com/recipes/2017/07/caesar-dressing-recipe.html")
```

```
r.content
```

```
soup = BeautifulSoup(r.content)
```

```
ingredients = soup.find('div', attrs = {'class': "recipe-ingredients"})
```

```
ingredient_list = [ingredient.text for ingredient in ingredients.find_all('li')]
```

```
ingredient_list
```

```
procedure = soup.find('div', attrs = {'class' : 'recipe-procedures'})
```

```
procedures = [procedure.text for procedure in procedure.find_all('p') if len(procedure) > 0]
```

```
procedures
```

```
summaries = soup.find('div', attrs = {'class' : 'recipe-introduction-body'})
```

```
summary = [summary.text for summary in summaries.find_all('p') if len(summary) > 0][1:]
```

```
summary
```

```
whys = soup.find('div', attrs = {'class' : 'recipe-why-body'})
```

```
why = [why.text for why in whys.find_all('li')]
```

```
why
```

```
header = "\n".join(summary + why)
```

```
print(header)
```

```
recipe = "\n".join(ingredient_list + [""] + procedures)
```

```
print (recipe)
```

```
formatted_recipe = "".join(header + "\n\n" + recipe)
```

```
print(formatted_recipe)
```
